## Tarefa (para agora)

1. Três séries, $v, y$ e $t$ seguem as seguintes regras:
        
\begin{align}
v[i] & = v[i-1] + 0,01 \times g \\
y[i] & = y[i-1] + 0,01 \times v[i-1] \\
t[i] & = t[i-1] + 0,01
\end{align}

Onde $g$ é a aceleração da gravidade, $g=-9,81 m/s^2$.
        
 a) Começando com v[0] = 10 m/s, y[0] = 0 m e t[0] = 0 s, calcule os 250 primeiros valores das três séries.

 b) Faça os gráficos com os valores encontrados de $y$ em função de $t$ e dos valores encontrados de $v$ em função de $t$.

Observação: o cálculo destas três séries é uma simulação computacional para calcular a posição e velocidade de uma partícula lançada na vertical com velocidade inicial de 10 m/s. O método utilizado para realizar esta simulação é conhecido como método de Euler.

```python
#a

v = [10]
y = [0]
t = [0]
g = -9.81
for i in range(1, 251):
    v.append(v[i -1] + 0.01*g)
    y.append(y[i -1] + 0.01*v[i - 1])
    t.append(t[i -1] + 0.01)

for j in range(250):
    print(f'v = {v[j]:.2f}, y = {y[j]:.2f}, t = {t[j]:.2f}')
```

```python
#b 1

import matplotlib.pyplot as plt

grafico = plt.plot(t, y)
```

```python
#b 2

import matplotlib.pyplot as plt

grafico = plt.plot(t, v)
```

2. O valor de $\ln(1+x)$ pode ser aproximado por (apenas para $|x|<1$):

$$ \ln(1+x) = x - \frac{x^2}{2} + \frac{x^3}{3} ... = \displaystyle\sum_{i=1}^n (-1)^{i+1}\frac{x^i}{i} $$

Onde $\ln$ é o logaritmo na base $e$ (conhecido como logaritmo natural).

 a) Use a expressão acima com $n = 100$ para encontrar $\ln(0.5)$.

 b) Compare o resultado com a função do Numpy `np.log(0.5)`.

```python
#a
# n = termo
# i = quando inicia

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt


n = 100
S = 0
x = -1/2
for i in range(1, n+1):
    S = S + (-1)**(i +1) * (x ** i) / i
print(f'S = {S}')
```
